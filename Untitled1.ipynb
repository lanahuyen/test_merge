```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import landscape, letter
from reportlab.pdfgen import canvas
from reportlab.platypus import Table, TableStyle, Paragraph
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors

# Load the data with UTF-8 encoding
file_path = '/Users/lanahuyen/UCB Bootcamp/Project 1/Group_Project_1/Resources/merged_df.csv'

data = pd.read_csv(file_path, encoding='utf-8')

# Ensure the Industry column exists
if 'Industry' not in data.columns:
    raise KeyError("'Industry' column not found in the dataset.")
```

```python
# Define the valid industries
valid_industries = ['Artificial intelligence', 'Auto & transportation', 'Consumer & retail','Cybersecurity', 'Data management & analytics',
                    'E-commerce & direct-to-consumer', 'Edtech', 'Hardware', 'Health', 'Internet', 'Internet software & services', 
                    'Mobile & telecommunications', 'Other', 'Supply chain, logistics, & delivery', 'Travel'
                ]  
```

```python
# Filter rows based on the valid industries
data = data[data['Industry'].isin(valid_industries)]

# Filter companies with complete data for all three years
data = data.dropna(subset=['2022_Valuation', '2023_Valuation', '2024_Valuation'])

# Clean Data
data['2022_Valuation'] = pd.to_numeric(data['2022_Valuation'].replace('[\$,]', '', regex=True))
data['2024_Valuation'] = pd.to_numeric(data['2024_Valuation'].replace('[\$,]', '', regex=True))

# Calculate Growth Rate
data['Growth'] = ((data['2024_Valuation'] - data['2022_Valuation']) / data['2022_Valuation']) * 100

# Drop bad data
data = data.dropna(subset=['Growth'])

# Save updated file
data.to_csv('/Users/lanahuyen/UCB Bootcamp/Project 1/Group_Project_1/Resources/updated_merged_df.csv', index=False)

data.head()
```

```python
print (data.columns)
```

```python
#  average growth per industry
industry_growth = data.groupby('Industry')['Growth'].mean().reset_index()

# Encode the Industry column using one-hot encoding to assign values to industry
data = pd.get_dummies(data, columns=['Industry'], drop_first=True)

# Industry columns after encoding
industry_cols = [col for col in data.columns if col.startswith('Industry_')]

# Perform regression analysis
X = sm.add_constant(data[industry_cols].astype(float))  # Add a constant term for the intercept
y = data['Growth'].astype(float)
model = sm.OLS(y, X).fit()

# Print the regression results
print("Growth Regression Results:")
print(model.summary())
```

```python
# Generate bar graph for average growth by industry
plt.figure(figsize=(10, 6))
plt.bar(industry_growth['Industry'], industry_growth['Growth'], color='skyblue')
plt.xlabel('Industry')
plt.ylabel('Average Growth (%)')
plt.title('Average Growth by Industry')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('average_growth_by_industry.png')
plt.show()
```

```python
# Generate heatmap of correlations
correlation_matrix = data[['Growth'] + industry_cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
# Generate residual plot
plt.figure(figsize=(10, 6))
residuals = y - model.fittedvalues
plt.scatter(model.fittedvalues, residuals, color='blue')
plt.axhline(y=0, color='red', linewidth=2)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.tight_layout()
plt.savefig('residual_plot.png')
plt.show()
```

```python
# Function to draw a table
def draw_table(data, col_widths, row_height, pdf, x_start, y_start):
    table = Table(data, colWidths=col_widths, rowHeights=row_height)
    style = TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
        ('GRID', (0, 0), (-1, -1), 1, colors.black),
    ])
    table.setStyle(style)
    table.wrapOn(pdf, x_start, y_start)
    table.drawOn(pdf, x_start, y_start)
```

```python
# Create the PDF in landscape mode
pdf = canvas.Canvas("analysis_summary.pdf", pagesize=landscape(letter))
width, height = landscape(letter)

# Title
pdf.setFont("Helvetica-Bold", 16)
pdf.drawString(30, height - 30, "Company Growth Analysis Summary")

# Analysis text
styles = getSampleStyleSheet()
text = """
The analysis shows the growth of companies within different industries from 2022 to 2024. The regression analysis helps us understand the impact of industry on company growth. 
The bar graph illustrates the average growth rate for each industry. 
The heatmap displays the correlation between different industries and growth, 
and the residual plot checks for random distribution of residuals, ensuring a good fit for the regression model. 
The regression analysis, which includes the one-hot encoded industry categories, indicates how different industries influence the growth rates of companies. 
Each dummy variable represents an industry and its coefficient shows the impact on growth compared to the baseline industry.
"""
paragraph = Paragraph(text, styles['Normal'])
paragraph.wrapOn(pdf, width - 60, height - 60)
paragraph.drawOn(pdf, 30, height - 120)

# Add the bar plot image
pdf.drawImage("average_growth_by_industry.png", 30, height - 270, width - 250, height - 500)

# Add the heatmap image
pdf.drawImage("correlation_heatmap.png", 30, height - 520, width - 250, height - 500)

# Add the residual plot image
pdf.drawImage("residual_plot.png", 30, height - 770, width - 250, height - 500)

# Regression Results Table
pdf.setFont("Helvetica-Bold", 14)
pdf.drawString(30, height - 820, "Growth Regression Results:")
regression_data = [["Variable", "Coefficient", "P-Value", "R_squared"]] + \
                  [["Intercept", f"{model.params[0]:.2f}", f"{model.pvalues[0]:.5f}", f"{model.rsquared:.2f}"]] + \
                  [[col, f"{model.params[i]:.2f}", f"{model.pvalues[i]:.5f}", ""] for i, col in enumerate(X.columns[1:], start=1)]
draw_table(regression_data, [100, 100, 100, 100], 20, pdf, 30, height - 920)

# Save the PDF
pdf.save()
print("PDF summary saved successfully.")
```
