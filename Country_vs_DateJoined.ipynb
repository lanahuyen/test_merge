```python
# Import Dependenciews
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
from pathlib import Path
```

```python
# Load and read CSV file
csvPath = Path("Project-1-CSVs/merged_df.csv")
mergedDF = pd.read_csv(csvPath, encoding="UTF-8")

# Drop original CSV indexing
mergedDF = mergedDF.drop(columns=['Unnamed: 0'])

# Display
mergedDF.head(10)
```

```python
# Reformat date Joined
mergedDF['Date Joined'] = pd.to_datetime(mergedDF['Date Joined'], format = '%m/%d/%Y')

# Create a sorted DataFrame
sortedMergedDF = pd.DataFrame(mergedDF)

# Sort companies from earliest join date to most recent join date
sortedMergedDF = sortedMergedDF.sort_values(by="Date Joined", ascending = True)

# Display top 10 (earliest joined)
sortedMergedDF.head(10)
```

```python
# Isolate the year from date
sortedMergedDF['Year Joined'] = sortedMergedDF['Date Joined'].dt.year

# Group by year and count number of companies joined per year
start_year = 2007
end_year = 2015
intervalOneFilteredDF = sortedMergedDF[(sortedMergedDF['Year Joined'] >= start_year) & (sortedMergedDF['Year Joined'] <= end_year)]
intervalOneFilteredDF
```

```python
intOneDF = intervalOneFilteredDF.groupby('Country').size().reset_index(name='Number of Unicorns')
print(intOneDF)

# Plotting the data
plt.figure(figsize=(10, 7))
plt.barh(intOneDF['Country'], intOneDF['Number of Unicorns'], color='seagreen')
plt.title(f'Number of Unicorn Companies by Country from {start_year} to {end_year}')
plt.xlabel('Number of Unicorns')
plt.ylabel('Country')
plt.grid(axis='x')
plt.show()
```

```python
# Group by year and count number of companies joined per year
startYear = 2016
endYear = 2022
intervalTwoFilteredDF = sortedMergedDF[(sortedMergedDF['Year Joined'] >= startYear) & (sortedMergedDF['Year Joined'] <= endYear)]
intervalTwoFilteredDF
```

```python
intTwoDF = intervalTwoFilteredDF.groupby('Country').size().reset_index(name='Number of Unicorns')

print(intTwoDF)

# Plotting the data
plt.figure(figsize=(25, 7))
# Plotting the data
plt.figure(figsize=(14, 8))  # Increase the figure size
bars = plt.barh(intTwoDF['Country'], intTwoDF['Number of Unicorns'], color='orange')
yTickPosition = range(len(intTwoDF))
# Add data labels
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()}', 
             va='center', ha='left')
plt.title(f'Number of Unicorn Companies by Country from {start_year} to {end_year}')
plt.xlabel('Number of Unicorns')
plt.ylabel('Country')
plt.yticks(yTickPosition, intTwoDF['Country'], fontsize=9)
plt.grid(axis='x')
plt.show()
```

Between 2007 and 2022, several significant global events and trends impacted the number and growth of unicorn startups. Here's a detailed look at some of these key factors:

Economic Factors
2008 Financial Crisis: The global financial crisis led to a severe economic downturn, reducing venture capital availability and making it harder for startups to secure funding. However, it also led to innovation in fintech as companies sought to address flaws in the traditional financial system.

Economic Recovery and Growth: Post-2010, many economies began to recover, leading to increased investment in technology and startups. This period saw a rise in venture capital availability and startup funding.

Low Interest Rates: Central banks, particularly the Federal Reserve, maintained low interest rates for much of this period, making borrowing cheaper and encouraging investment in high-growth startups.

Technological Advancements
Smartphone Revolution: The launch of the iPhone in 2007 and subsequent smartphone proliferation revolutionized mobile computing, creating opportunities for numerous app-based startups.

Cloud Computing: The growth of cloud computing services (e.g., AWS, Microsoft Azure) reduced the cost of starting and scaling tech companies, enabling more startups to achieve unicorn status.

Big Data and AI: Advances in artificial intelligence and big data analytics opened up new avenues for innovation, leading to the rise of AI-driven startups.

Social and Demographic Changes
Changing Consumer Behavior: The rise of social media, e-commerce, and the gig economy shifted consumer behavior, creating new markets for startups in sectors like digital marketing, online retail, and ride-sharing.

Urbanization and Population Growth: Rapid urbanization and growing middle-class populations in emerging markets expanded the customer base for many tech startups.

Political and Regulatory Environment
Startup Ecosystem Development: Many countries and regions developed policies to support startups, including tax incentives, reduced regulatory burdens, and innovation hubs (e.g., Silicon Valley, Shenzhen).

GDPR and Data Privacy Regulations: The implementation of data privacy regulations like the EU's GDPR in 2018 impacted how startups handle user data, creating both challenges and opportunities.

Global Events
COVID-19 Pandemic: The pandemic accelerated digital transformation, increasing demand for remote work tools, e-commerce, digital health, and other tech solutions. This period saw a surge in unicorn startups in these sectors.

Geopolitical Tensions: Trade wars, particularly between the US and China, affected global supply chains and created uncertainty, impacting startups dependent on international trade.

Cultural Factors
Rise of the Entrepreneurial Culture: Increased media coverage and success stories of startups achieving unicorn status inspired more entrepreneurs, leading to a vibrant startup culture globally.

Access to Talent: The globalization of talent, facilitated by remote work and international education, allowed startups to access skilled professionals from around the world.

Examples of Key Changes and Their Impact
2008 Financial Crisis: Despite initial challenges, the crisis led to innovations in fintech, with companies like Square (founded in 2009) emerging to address issues in the financial sector.

Growth of Social Media: Platforms like Facebook (went public in 2012) and Instagram (acquired by Facebook in 2012) created new opportunities for digital marketing and social media startups.

Cloud Computing Boom: The widespread adoption of cloud services allowed companies like Airbnb (founded in 2008) and Slack (launched in 2013) to scale rapidly.

AI and Machine Learning: Startups like UiPath (founded in 2005 but gained prominence in the 2010s) leveraged AI and automation to become unicorns.

COVID-19 Pandemic: Companies like Zoom (founded in 2011, became a unicorn in 2017) saw explosive growth due to the shift to remote work, highlighting the impact of global events on startup success.

By analyzing these factors, one can understand the complex interplay of economic, technological, social, political, and global influences on the growth of unicorn startups between 2007 and 2022.
